```python
# You code here (Please add comments in the code):
# Import necessary libraries
from bs4 import BeautifulSoup
import requests
import pandas
title = []
fullReview =[]
# Loop to scrape data from multiple pages (70 in this case)
for i in range(70):
  URL = "https://www.flipkart.com/apple-2020-macbook-air-m1-8-gb-256-gb-ssd-mac-os-big-sur-mgn63hn-a/p/itm3c872f9e67bc6?pid=COMFXEKMGNHZYFH9&lid=LSTCOMFXEKMGNHZYFH9P56X45&marketplace=FLIPKART&q=macbook&store=6bo%2Fb5g&srno=s_1_1&otracker=search&otracker1=search&fm=search-autosuggest&iid=b5a555c9-768c-4559-bfe9-446fca83be8d.COMFXEKMGNHZYFH9.SEARCH&ppt=sp&ppn=sp&ssid=to0qougjg00000001694919311787&qH=864faee128623e2f" + str(i)
  webPage = requests.get(URL)   # Send an HTTP GET request to the URL
  beautifulSoup = BeautifulSoup(webPage.text, 'html.parser')   # Parse the HTML content of the webpage using BeautifulSoup
  title_text = beautifulSoup.find_all(class_='_2-N8zT')     # Find all elements with class '_2-N8zT' (presumably titles)
  fullReview_text = beautifulSoup.find_all(class_='t-ZTKy')
   # Iterate through pairs of title and full review elements and append them to the respective lists
  for ele1, ele2 in zip(title_text, fullReview_text):
      title.append(ele1.text)
      fullReview.append(ele2.text) # Create a DataFrame to store the scraped data
res = pandas.DataFrame(list(zip(title, fullReview)), columns =['Title', 'Full text review'])
res


```

Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2013-2023).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
import requests
from bs4 import BeautifulSoup

# Function to scrape and print articles from a given URL
def scrape_articles(url):
    response = requests.get(url)  # Send a GET request to the URL
    soup = BeautifulSoup(response.text, 'html.parser')   # Use BeautifulSoup to extract the page's HTML content.
    articles = soup.find_all('div', class_='gs_ri')  # Locate and gather data for each article on the page.

  # Loop through the articles and extract the desired information
    for article in articles:
        title = article.find('h3', class_='gs_rt').text
        venue = article.find('div', class_='gs_a').text.split(' - ')[-1]
        year = article.find('div', class_='gs_a').text.split(' - ')[0]
        authors = article.find('div', class_='gs_a').text.split(' - ')[1]
        abstract = article.find('div', class_='gs_rs').text

        print("Title:", title)
        print("Venue:", venue)
        print("Year:", year)
        print("Authors:", authors)
        print("Abstract:", abstract)
        print("\n---\n")

# Loop through plenty of pages of results to get 1,000 articles.
articles_per_page = 10  # A page of Google Scholar normally shows 10 items.
total_articles = 1000

for page in range(0, total_articles, articles_per_page):
  # URL of Google Scholar search results
    page_url = f"https://scholar.google.com/scholar?hl=en&as_sdt=0%2C44&q=computer+science&btnG=&oq=computer+{page}"
    scrape_articles(page_url)
```
